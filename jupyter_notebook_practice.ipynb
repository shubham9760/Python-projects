```python
#Revenue generated by room category
df.groupby("room_category")["revenue_generated"].sum().plot(x="room_category", y="revenue_generated", kind="bar")
plt.title("Category wise revenue generated")
plt.xlabel("room_category")
plt.ylabel("revenue_generated")
plt.show()
```

```python
#Revenue generated by city
df.groupby("city")["revenue_generated"].sum().plot(x="city", y="revenue_generated", kind="line")
plt.xlabel("city")
plt.ylabel("revenue_generated")
plt.title("Revenue generated by city")
plt.show()
```

```python
#Revenue generated by day type
df.groupby("day_type")["revenue_generated"].sum().plot(x="day_type", y="revenue_generated", kind="bar")
plt.xlabel("day_type")
plt.ylabel("revenue_generated")
plt.title("Revenue generated by day_type")
plt.show()
```

```python
#Revenue realized by booking status
df.groupby("booking_status")["revenue_realized"].sum().plot(kind="line")
plt.xlabel("booking_status")
plt.ylabel("revenue_realized")
plt.title("Revenue realized by booking status")
plt.show()
```

```python
#Revenue realized by room class
df.groupby("room_class")["revenue_realized"].sum().plot(x="room_class", y="revenue_realized", kind="bar")
plt.xlabel("room_class")
plt.ylabel("revenue_realized")
plt.title("Revenue realized by room class")
plt.show()
```

```python
#Revenue realized by booking platform
df.groupby("booking_platform")["revenue_realized"].sum().plot.pie(y="revenue_realized", autopct="%1.1f%%")
plt.xlabel("booking platform")
plt.ylabel("revenue_realized")
plt.title("Revenue realized by booking platform")
plt.show()
```

```python
#Number of guest by room category
df.groupby("room_category")["no_guests"].sum().plot.bar(y="no_guest")
plt.xlabel("room_category")
plt.ylabel("no_guest")
plt.title("Number of guest by room category")
plt.show()
```

```python
#Average occupancy by room category
df_fact_aggregated_bookings.groupby("room_category")["occ_pct"].mean().plot.bar(y="occ_pct")
plt.title('Average Occupancy Per Room Category')
plt.xlabel("room_category")
plt.ylabel("occ_pct")
plt.show()
```
