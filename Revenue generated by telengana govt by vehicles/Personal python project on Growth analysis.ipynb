**How does the revenue generated from document registration vary 
across districts in Telangana? List down the top 5 districts that showed 
the highest document registration revenue growth between FY 2019 
and 2022**

```python
df_revenue_generated = df_fact_stamps.merge(df_dim_districts, how="left", on="dist_code")
df_revenue_generated
```

```python
df_revenue_generated = df_revenue_generated.merge(df_dim_date, how="left", on="month")
df_revenue_generated
```

```python
df_revenue_generated = df_revenue_generated[df_revenue_generated["fiscal_year"] >= 2019]
df_revenue_generated
```

```python
revenue_generated_by_registered_document = df_revenue_generated.groupby("district")["documents_registered_rev"].sum().sort_values(ascending=False)
revenue_generated_by_registered_document.iloc[0:5]
```

**How does the revenue generated from document registration compare 
to the revenue generated from e-stamp challans across districts? List 
down the top 5 districts where e-stamps revenue contributes 
significantly more to the revenue than the documents in FY 2022?**

```python
revenue_generated_by_estamps = df_revenue_generated[df_revenue_generated["fiscal_year"]==2022].groupby("district")["estamps_challans_rev"].sum().sort_values(ascending=False)
revenue_generated_by_estamps.head(5)
```

```python
revenue_generated_by_registered_document.head(5)
```

```python
revenue_generated_by_registered_document.info()
```

```python
revenue_generated_by_estamps.info()
```

```python
df_revenue_generated["estamp_VS_registered_rev"] =np.where((df_revenue_generated["estamps_challans_rev"]>df_revenue_generated["documents_registered_rev"]),df_revenue_generated["estamps_challans_rev"], np.nan)
df_revenue_generated
```

```python
df_revenue_generated = df_revenue_generated[df_revenue_generated["fiscal_year"]==2022].sort_values(by="estamp_VS_registered_rev", ascending=False)
df_revenue_generated
```

sector investments during FY 2019 to 2022?**
